# Plotte andregradsfunksjonar

Skal laga eit program som plottar funksjonar
$$f(x)=ax^2+bx+c$$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, a, b, c):
    return a*x**2 + b*x + c

# INPUT
a = 2
b = -2
c = -3
start = -3
slutt = 3

# x-verdiar
x = np.linspace(start, slutt, 100)

# y-verdiar
y = f(x, a, b, c)

# plotting
plt.plot(x, y, label="f(x)")

# pynting
plt.grid()
plt.axhline(0, c="black")
plt.axvline(0, c="black")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafen til f(x)")
plt.show()
```

## Utvider

Legg til plotting av symmetrilinje og ekstremalpunkt som ekstrafunksjon.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, a, b, c):
    return a*x**2 + b*x + c

# INPUT - endra desse for  passa til funksjonen
a = 1
b = 0
c = -25
start = -8
slutt = 8
ekstra = True # viss True: teikne symmetri + ekstremal.

# x-verdiar
x = np.linspace(start, slutt, 100)

# y-verdiar
y = f(x, a, b, c)

# plotting
plt.plot(x, y, label="f(x)")

# ekstremalpunkt og symmetrilinje
if ekstra:
    sym = (-b)/(2*a)
    y_sym = f(sym, a, b, c)
    plt.plot(sym, y_sym, 
             "o", 
             label="ekstremalpunkt", 
             c = "red"
            )
    plt.axvline(sym,                     # xverdi
                c = "grey",              # farge
                label="symmetrilinje",   # merkelapp
                ls = "--"                # linjestil
               )

# pynting
plt.grid()
plt.axhline(0, c="black")
plt.axvline(0, c="black")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafen til f(x)")
plt.show()
```
