# Program som lyser andregradslikningar

Algoritmen illustrert i eit flytskjema:


```python
import numpy as np

print("Program som lyser likninga ax^2+bx+c=0")

a = float(input("Skriv inn a: "))
b = float(input("Skriv inn b: "))
c = float(input("Skriv inn c: "))

# rekne ut diskriminanten
d = b**2 - 4*a*c

# sjekke kor mange lysingar me har
if d > 0:
    # to lysingar
    x1 = (-b-np.sqrt(d))/2*a
    x2 = (-b+np.sqrt(d))/2*a
    print("To lysingar")
    print(f"x1 = {x1}, x2 = {x2}")
    
elif d == 0: 
    # ei lysing
    x1 = (-b)/2*a
    print("Ei lysing")
    print(f"x1 = x2 = {x1}")
    
else: 
    # inga lysing
    print("Inga lysing. d < 0")
```

## For spesielt interesserte

Lagar "andregradslikningslysar" ved  definera ein funksjon. Testar p likningane 
$$
    \begin{align}
        x^2+6x+5&=0 \\ 
        x^2+6x+9&=0 \\
        x^2+6x+11&=0
    \end{align}
$$

```python
def abc_formel(a, b, c):
    import numpy as np
    
    # rekne ut diskriminanten
    d = b**2 - 4*a*c

    # sjekke kor mange lysingar me har
    if d > 0: # to lysingar
        x1 = (-b-np.sqrt(d))/2*a
        x2 = (-b+np.sqrt(d))/2*a
        return f"To lysingar \nx1 = {x1}, x2 = {x2}\n"
    elif d == 0: # ei lysing
        x1 = (-b)/2*a
        return f"Ei lysing \nx1 = x2 = {x1}\n"
    else: # inga lysing
        return "Inga lysing. d < 0 \n"

    
print("likning 1:")
print(abc_formel(1, 6, 5))

print("likning 2")
print(abc_formel(1, 6, 9))

print("likning 3")
print(abc_formel(1, 6, 11))
```
