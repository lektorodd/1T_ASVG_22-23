# Plotte funksjonar

Generell plotting av funksjonar i Python. 

1) Definere funksjonen
2) Lage x-verdiar
3) Rekne ut y-verdiar
4) Plotte funksjonen
5) Pynte plottet

## DÃ¸me:

```python
# (1) definere funksjonen
def g(x):
    return 2*1.05**x

# (2) lage x-verdiar
x = np.linspace(0, 20, 100)

# (3) rekne ut y-verdiar
y = g(x)

# (4) plotte
plt.plot(x,y, label="g(x)")

# (5) pynting
plt.grid()
#plt.axhline(0, c="black")
#plt.axvline(0, c="black")
plt.ylim([0,6])
plt.xlim([0,21])
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
